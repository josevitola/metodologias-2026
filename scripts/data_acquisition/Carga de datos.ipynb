```python
from pathlib import Path

DATASET_PATH = Path("/content/dataset/dataset_preparado")
```

```python
import os

print(os.listdir(DATASET_PATH))
```

Calculamos el tamaño y los formatos presentes en el dataset:

```python
import os
from collections import Counter

total_files = 0
format_count = Counter()
total_size_bytes = 0

for root, dirs, files in os.walk(DATASET_PATH):
    for file in files:
        file_path = Path(root) / file

        total_files += 1

        ext = file_path.suffix.lower()
        format_count[ext] += 1

        total_size_bytes += file_path.stat().st_size

total_size_mb = total_size_bytes / (1024 * 1024)

print("===== RESUMEN DATASET =====")
print(f"Total de documentos (imágenes): {total_files}")
print("\nFormatos encontrados:")
for fmt, count in format_count.items():
    print(f"{fmt}: {count}")

print(f"\nTamaño total del dataset: {total_size_mb:.2f} MB")
```

```python
person_counts = {}

for person_folder in DATASET_PATH.iterdir():
    if person_folder.is_dir():
        count = len([f for f in person_folder.iterdir() if f.is_file()])
        person_counts[person_folder.name] = count

print("===== IMÁGENES POR PARTICIPANTE =====")
for person, count in person_counts.items():
    print(f"{person}: {count}")
```

El conjunto de datos cargado está compuesto por imágenes correspondientes a los integrantes del grupo. Estas fotografías constituyen la entrada inicial del proyecto de reconocimiento facial y se encuentran organizadas por participante.
